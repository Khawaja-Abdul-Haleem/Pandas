## Line Plot

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [10,20,30,40,30,25]
plt.plot(x,y , color = 'red')
#plt.figure(figsize=(2,5))
plt.xlabel('X AXIS', fontsize = 20)
plt.ylabel('Y AXIS', fontsize = 20)
plt.title("LINE Plot", fontsize = 20)
plt.show()
```

## Bar Graph

```python
y1 = [1,2,3,4,5,6]
y2 = [10,20,30,15,25,5]
x = ['English', 'Maths', 'Science', 'Computer', 'Biology', 'Social Study']
plt.bar(y1,y2,tick_label = x,width = 0.5, color = ['red', 'Green', "Blue", 'grey', 'Yellow', 'purple'])
#plt.figure(figsize=(2,5))
plt.xlabel('X AXIS', fontsize = 20)
plt.ylabel('Y AXIS', fontsize = 20)
plt.title("Bar Graph", fontsize = 20)
plt.show()
```

## Histogram

```python
list = [10,2,4,5,6,7,33,3,10,1,3,5,6,7,2,9]
range = (0,50)
bins = 10
plt.hist(list,bins,range)
plt.xlabel('X AXIS', fontsize = 20)
plt.ylabel('Y AXIS', fontsize = 20)
plt.title("Histogram", fontsize = 20)
plt.show()
plt.show()
```

## Scatter Plot

```python
x = [1,2,3,4,5,6]
y = [10,20,30,40,30,25]
plt.scatter(x,y , color = 'green', marker = '*', s = 150, label = 'Stars')
#plt.figure(figsize=(2,5))
plt.xlabel('X AXIS', fontsize = 20)
plt.ylabel('Y AXIS', fontsize = 20)
plt.title("Scatter Plot", fontsize = 20)
plt.legend(loc = 2)
plt.show()
```

## Pie Chart

```python
slices = [10,2,4,10,45,20]
activities = ['Sleep', 'Eat', 'Play', 'Game', 'Mobile', 'Study']
color = ['red', 'Green', "Blue", 'grey', 'Yellow', 'purple']
plt.pie(slices, labels = activities, colors = color, explode = (0,0,0.5,0,0,0.2), autopct = '%1.2f%%')
plt.show()
```
