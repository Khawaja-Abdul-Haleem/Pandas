# Handling Missing Data

For Numeric data, pandas uses the floating-point value NAN to represent missing data

```python
import pandas as pd
from numpy import nan as NA

data = pd.DataFrame([[1., 6.5, NA], [1., NA, NA],
                    [NA, NA, NA], [NA, 6.5, NA]], 
                   columns = list('abc'))
#print(data)
display(data)


cleaned = data.dropna(how = 'all') # remove rows that have all null values
display(cleaned) 

cleaned_1 = data.dropna() # remove row that has atleast one null values
display(cleaned_1)

# For Columns-Wise
cleaned_col = data.dropna(axis = 1, how = 'all')
display(cleaned_col)
```

# Filling Missing Value

```python
import numpy as np

df = pd.DataFrame(np.random.randn(7,3))
df.iloc[:4, 1] = NA
df.iloc[:2, 2] = NA
display(df)

display(df.fillna(1.5)) # replace NaN value with 1.5

display(df.fillna({1: 0.5, 2: 0})) #columns Number_1 filled with 0.5 and column number_2 filled with 0

# fill not values from left column (axis = 0 means from top row)
display(df.fillna(method = 'ffill', axis = 1))

df.fillna(method = 'ffill', limit = 1, axis = 1)
```

# Removing Duplicates

```python
import pandas as pd

data = pd.DataFrame({'k1': ['one', 'two']*3 + ['two'], 'k2' : [1,1,2,3,3,4,4]})

display(data)
display(data.duplicated()) # remove when all data is duplicated
display(data.drop_duplicates())

data['v1'] = ['one', 'two', 'one', 'four', 'one', 'six', 'two']
display(data)

data = data.drop_duplicates(['k1', 'v1'])
data
```

# Combining And Merging DataSets

<ul>
    <li>pandas.merge connects rows in DataFrames based on one or more keys.</li>
    <li>pandas.concat concatenates or "stacks" together object along an axis </li>
    <li>The Combine_first instance method enables splicing together overlaping data to fill in missing values in one object with value from another</li>


```python
df_1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data1' : range(7)})

df_2 = pd.DataFrame({'key' : ['a', 'b', 'd'],
                    'data_2' : range(3)})

display(df_1)
display(df_2)
pd.merge(df_1, df_2, on = "key")
```

```python
df_3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'data_1' : range(7)})

df_4 = pd.DataFrame({'rkey' : ['a', 'b', 'd'],
                    'data_2' : range(3)})

display(df_3)
display(df_4)

pd.merge(df_3, df_4, left_on = 'lkey', right_on = 'rkey')
```

```python
df_4 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                    'value' : range(7)})

df_5 = pd.DataFrame({'group_val' : [3.5, 7]},
                    index = ['a', 'b'])

display(df_4)
display(df_5)

display(pd.merge(df_4, df_5, left_on = "key", right_index = True)) # inner join
display(pd.merge(df_4, df_5, left_on = "key", right_index = True, how = 'outer'))
display(pd.merge(df_4, df_5, left_on = "key", right_index = True))

```

# Concatenating Along an Axis
<ul><li>Another Kind of data Combination operation is referred to interchangeably as concatenation, binding or Stacking</li></ul>

```python
s1 = pd.Series([0,1], index = ['a', 'b'])
s2 = pd.Series([2, 3, 4], index = ['c', 'd', 'e'])
s3 = pd.Series([5,6], index = ['f', 'g'])

display(s1)
display(s2)
display(s3)

display(pd.concat([s1, s2, s3])) # by default concat works on axis = 0

pd.concat([s1,s2,s3], axis = 1)
# In this case, there is no overlap on other axis, which as u can see is sorted 
# union(the 'outer' join ) of indexes. You can instead intersect them by passing join = 'inner'
```

```python
s4 = pd.concat([s1, s3])
display(pd.concat([s1, s4], axis = 1))
display(pd.concat([s1, s4], axis = 1, join = 'inner'))
```
