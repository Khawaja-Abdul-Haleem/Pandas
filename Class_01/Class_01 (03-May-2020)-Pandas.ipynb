### You can replace index value in series data

```python
obj1 = pd.Series([1,2,3,4,5], index=["A",'B',"C","D","E"])
obj1
```

```python
obj1 = pd.Series([1,2,3,4,5], index=["A",'B',"C","D","E"])
obj1.index.name = "CNIC"
obj1
```

```python
obj2 = pd.Series([8,9,3], index=["A","D","Z"])
obj2
```

```python
print(obj1,obj2,sep="\n")
obj1 + obj2
```
