# pd.concat

```python
pd.concat([df1, df2], axis=0)
```

```python
df = pd.concat([df1, df2], axis=1)
df
```

```python
df['total'] = df.A + df.B + df.C + df.D + df.E
df['percentage'] = (df.A + df.B + df.C + df.D + df.E)/500*100
def grade(p):
    if p>=80:
        return "A+"
    elif p>=70:
        return "A"
    elif p>=60:
        return "B"
    elif p>=50:
        return "C"
    else:
        return "Fail"
df['grade'] = df.percentage.apply(grade)
df
```

```python
df[["Course","A","B","C","D","E"]]
```

```python
def user_fn(course, c1, c2, c3, c4, c5):
    if course == "AI":
        return np.array([c1,c2,c3,c4,c5]).sum()
    elif course == "AIOT":
        return np.array([c1,c2,c3,c4,c5]).prod()
    elif course == "CNC":
        return sum(np.array([c1,c2,c3,c4,c5]).cumsum())
```

```python
np.array([1,2,3]).sum()
```

```python
df["CheckValidation"]=df[["Course","A","B","C","D","E"]].apply(lambda x:user_fn(*x), axis=1)
df
```
